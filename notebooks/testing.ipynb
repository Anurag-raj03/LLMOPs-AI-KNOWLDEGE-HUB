```python
import json 
import pandas as pd
data=[]
with open("E:/LLMOPs/aiops/data/raw/arxiv-metadata-oai-snapshot.json", "r") as f:
    for i,line in enumerate(f):
        if i>= 100000:
            break
        data.append(json.loads(line))
print("Loaded: ",len(data))
print(data[0].keys())        
```

```python
import pandas as pd

df = pd.DataFrame(data)
df = df[["id", "title", "categories", "abstract"]].dropna()
df.head()
```

```python
df["text"]=df["title"].str.strip()+". "+df["abstract"].str.strip()
df=df[["id","categories","text"]]
```

```python
subset_df=df.sample(n=20000,random_state=42)
subset_df.reset_index(drop=True,inplace=True)
subset_df.shape
```
